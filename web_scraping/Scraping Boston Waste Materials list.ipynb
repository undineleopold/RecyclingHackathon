```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys  # select Keys and hit F4 to see the constants available
import time
import pandas as pd
```

```python
driver = webdriver.Firefox()
driver.get("https://www.boston.gov/departments/public-works/recycling-boston#!rc-cpage=wizard_material_list")
time.sleep(10) #need to wait at least 10 seconds for the widget to load
#elem= driver.find_element(By.CSS_SELECTOR, ".rCpage label") #try locating the text above the search box
#print(elem.text) #success: "TYPE THE NAME OF AN ITEM AND WE'LL TELL YOU HOW TO RECYCLE OR DISPOSE OF IT."
elems = driver.find_elements(By.CSS_SELECTOR, ".rCpage h4 a") #try locating the clickable letters of the alphabet on the list
rclist=[]
for elem in elems:
    print(elem.text)
    elem.click() #expand letter menu
    time.sleep(3)
    categories=driver.find_elements(By.CSS_SELECTOR, ".rCpage ul a")
    for category in categories:
        rclist.append([category.text, category.get_attribute("href")]) 
    elem.click() #unexpand letter
    time.sleep(2)
#elem = driver.find_element(By.CSS_SELECTOR, ".rC-left-link a") #. is for class name
#print(elem.text) #the link "List of Materials" below the wizard
driver.close()
```

```python
rc_links=pd.DataFrame(rclist, columns=['material','link'])
rc_links.head(50)
```

```python
#okay, make a third column with just the last 5 digits in the link
rc_links['page_id']=rc_links['link'].apply(lambda x: x[-5:])
rc_links.head(50)
```
